```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import lightgbm as lgb
```

```python
dataset = pd.read_csv("supplement.csv")
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset["Store_Type"] = dataset["Store_Type"].map({"S1": 1, "S2": 2, "S3": 3, "S4": 4})
dataset["Location_Type"] = dataset["Location_Type"].map({"L1": 1, "L2": 2, "L3": 3, "L4": 4, "L5": 5})
dataset["Region_Code"] = dataset["Region_Code"].map({"R1": 1, "R2": 2, "R3": 3})
dataset["Discount"] = dataset["Discount"].map({"Yes": 1, "No": 0})
```

```python
dataset["Store_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

```python
dataset["Discount"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```
